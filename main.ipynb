```python
import pandas as pd
```

```python
df = pd.read_csv('2023to2025 interviews.CSV - 2023to2025 interviews.CSV.csv', encoding='utf-8-sig')
```

```python
extract_pattern = r'^[Qq]*(\d+\s*[-]*\s*\d*)\s*[\.\)]'
df['question_number'] = df['question'].str.extract(extract_pattern, expand=False)
df['question'] = df['question'].str.replace(extract_pattern, '', regex=True)
df['question'] = df['question'].str.strip()
```

```python
df.iloc[60:100, :]
```

```python
df['parent'] = None

for i, row in df.iterrows():
    qnum = row['question_number']
    if isinstance(qnum, str) and '-' in qnum:
        parent_num = qnum.split('-')[0].strip()
        parent_idx = df[
            (df['question_number'] == parent_num) &
            (df['category'] == row['category']) &
            (df['company'] == row['company']) &
            (df['question_at'] == row['question_at'])
        ].index
        if len(parent_idx) > 0:
            df.at[i, 'parent'] = parent_idx[0]

parent = [0 for _ in range(10)]

for i, question in enumerate(df['question']):
    level = question.count('-') + question.count('ㄴ')
    if level > 0:
        df.at[i, 'parent'] = parent[level - 1]
        df.at[i, 'question'] = question.replace('-', '').replace('ㄴ', '').strip()
    parent[level] = i
```
